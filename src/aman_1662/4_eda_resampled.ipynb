```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
```

### Subset Dataset

```python
# Step 1: Load the dataset
data = pd.read_csv('./data/subset_dataset_resampled.csv', parse_dates=['Datetime'], index_col='Datetime')

# Step 2: Data Cleaning

# Check for missing values
print("Missing values per column:")
print(data.isnull().sum())

# Step 3: Summary Statistics
print("\nSummary Statistics:")
print(data.describe())

# Step 4: Time Series Analysis

# Plot the time series for all columns
plt.figure(figsize=(12, 8))
data.plot(subplots=True, layout=(4, 2), figsize=(12, 12), sharex=True)
plt.tight_layout()
plt.show()

# Resample the data (if needed)
daily_data = data.resample('D').mean()  # Daily average
weekly_data = data.resample('W').mean()  # Weekly average
monthly_data = data.resample('M').mean()  # Monthly average

# Plot the resampled data (daily in this case)
plt.figure(figsize=(12, 6))
daily_data['Global_active_power'].plot(title='Daily Global Active Power')
plt.show()

# Step 5: Correlation Analysis
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()

# Step 6: Trend, Seasonality, and Noise Decomposition

# Decompose the Global Active Power time series to check trend and seasonality
result = seasonal_decompose(data['Global_active_power'], model='additive', period=365)  # Adjust period as necessary
result.plot()
plt.show()

# Optional: Check for outliers in 'Global_active_power'
plt.figure(figsize=(12, 6))
sns.boxplot(x=data['Global_active_power'])
plt.title('Boxplot of Global Active Power')
plt.show()
```
